```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
```

```python
Data = pd.read_csv('netflix titles.csv', error_bad_lines=False)
```

```python
Data
```

```python
a = Data[['type', 'title', 'country', 'release_year', 'rating', 'listed_in']]
a
```

```python
b = a.groupby('type')
b['type'].size()
```

```python
c = a[a.type == 'Movie']
c.reset_index(drop=True)
```

```python
d = c.groupby('country')
d['country'].size()
```

```python
e = c[c.country == 'United States']
e.reset_index(drop=True)
```

```python
f = e.groupby('release_year')
f['release_year'].size()
```

```python
g = e[e.release_year == '2020']
g.reset_index(drop=True)
```

```python
h = g.groupby('rating')
h['release_year'].size()
```

```python
i = g[g.rating == 'TV-MA']
i.reset_index(drop=True)
```

```python
j = i.groupby('listed_in')
j['release_year'].size()
```

```python
k = i[i.listed_in == 'Documentaries, Music & Musicals']
k.reset_index(drop=True)
```
